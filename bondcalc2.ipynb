```python
df = pd.read_csv('lindarollover2.csv')
#df["Expiration/Maturity"]
df2=df[["Symbol","Description","Quantity","Expiration/Maturity"]]
df2

df3=df2.sort_values(by=["Expiration/Maturity"],ascending=False)
df3
```

```python
df4=df3[df3["Expiration/Maturity"]!="--"]
df4
```

```python
from datetime import datetime
#df4["maturity"] = pd.to_datetime(df4["Expiration/Maturity"])

df4.loc[:,("maturity")] = pd.to_datetime(df4["Expiration/Maturity"])
df4.loc[:,("Quantity")] = df4["Quantity"].str.replace(',','')
#df4.groupby(by=['maturityq']).agg(['count'])
df5=df4[["maturity","Quantity"]]
df5
#df5.groupby(by=['maturityq']).agg(['sum'])
```

```python
df5.loc[:,("maturityq")] = pd.PeriodIndex(df4["maturity"], freq='Q')
d=df5.sort_values(by=["maturity"],ascending=True)
d.loc[:,("Quantityi")] = df4["Quantity"].astype(int)
df5 = d[["maturity","Quantityi"]]
df5
```
